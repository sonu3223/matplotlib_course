# savefig

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,3,4,9,8]
plt.plot(x,y,color="g")
plt.savefig("lineftf.pdf") #jpg png
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,3,4,9,8]
plt.plot(x,y,color="g")
plt.savefig("lin12",dpi=2000,facecolor="r",transparent=False,bbox_inches="tight")
plt.show()
```
