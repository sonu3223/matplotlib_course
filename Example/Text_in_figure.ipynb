```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,3,4,9,8]
plt.plot(x,y,color="g")

plt.title("plot",fontsize=15,color="b")
plt.xlabel("day",fontsize=15,color="r")
plt.ylabel("no.",fontsize=15,color="r")

plt.text(3,4,"java",fontsize=15,color="y",style="italic",bbox={"facecolor":"g"})

plt.annotate("python",xy=(2,3),xytext=(2,5),arrowprops=dict(facecolor="r",shrink=100))
plt.legend(["up"],loc=2,facecolor="r",edgecolor="y",framealpha=.2,shadow=True)
plt.show()
```
