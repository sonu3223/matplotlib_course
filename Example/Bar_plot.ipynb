# Bar plot

```python
import matplotlib.pyplot as plt
x=["python","c","c++","java"]
y=[90,70,60,82]
plt.bar(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=["python","c","c++","java"]
y=[90,70,60,82]
plt.xlabel("Language",fontsize=10)
plt.ylabel("No.",fontsize=10)
plt.title("Data",fontsize=10)
c=["r","y","g","b"]
plt.bar(x,y,color=c) #align=center,edge  ,width=0.8, edgecolor="black",  linewidth=2,  linestyle=":",  alpha=.9, label="Dataplot"
plt.legend("Dataplot")
plt.show()
```

# multiple barplot

```python
import matplotlib.pyplot as plt
x=["python","c","c++","java"]
y=[90,70,60,82]
z=[20,30,40,50] #for other barplot
plt.xlabel("Language",fontsize=20)
plt.ylabel("No.",fontsize=20)
plt.title("Data",fontsize=20)
plt.bar(x,y,color="r",label="Dataplot")
plt.bar(x,z,color="b",label="Dataplot1") #another barplot
plt.legend()
plt.show()
```

# multiple barplot showside(not over lap)

```python
import matplotlib.pyplot as plt
import numpy as np
x1=["python","c","c++","java"]
y1=[90,70,60,82]
z1=[20,30,40,50]

width=0.2
p=np.arange(len(x1))
p1=[j+width for j in p]

# width=0.2
plt.xlabel("Language",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.title("Data",fontsize=15)

plt.bar(p,y,color="r",label="Dataplot")
plt.bar(p1,z,color="g",label="Dataplot1") 

plt.xticks(p+width/2,x,rotation=10)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
x=["python","c","c++","java"]
y=[90,70,60,82]
z=[20,30,40,50] 

plt.xlabel("Language",fontsize=20)
plt.ylabel("No.",fontsize=20)
plt.title("Data",fontsize=20)

plt.barh(x,y,color="r",label="Dataplot")
plt.barh(x,z,color="b",label="Dataplot1")

plt.legend()
plt.show()
```
