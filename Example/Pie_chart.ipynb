# Pie plot

```python
import matplotlib.pyplot as plt
x=[10,20,30,40,50,60]
plt.pie(x)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40,50,60]
y=["c","c++","python","java","sql","npl"]
plt.pie(x,labels=y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40,50,60]
y=["c","c++","python","java","sql","npl"]
ex=[0.0,0.3,0.0,0.0,0.0,0.0]
c=["y","b","g","r","orange","pink"]
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%",shadow=True,radius=1.0,labeldistance=1.1,startangle=0,
        textprops={"fontsize":15},counterclock=True,wedgeprops={"linewidth":2,"edgecolor":"r"},rotatelabels=False)
#"width":0.1 in wedgeprops,  center=(3,3)
plt.title("PIE PLOT")
plt.legend(loc=2)
plt.show()
```

# donate(multiple) pie chard

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50,60]
x1=[30,20,40,10,60,10]
y=["c","c++","python","java","sql","npl"]
c=["y","b","g","r","orange","pink"]
plt.pie(x,labels=y,radius=1.5)
plt.pie(x1,radius=.5,colors=c)
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50,60]
y=["c","c++","python","java","sql","npl"]
plt.pie(x,labels=y,radius=1.5)
plt.pie([1],colors="b")
plt.show()
```

```python

x=[10,20,30,40,50,60]
y=["c","c++","python","java","sql","npl"]
plt.pie(x,labels=y,radius=1.5)
c=plt.Circle(xy=(0,0),radius=1,facecolor="w")
plt.gca().add_artist(c)
plt.show()
```
