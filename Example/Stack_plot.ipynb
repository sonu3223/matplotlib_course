```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
area=[6,2,4,5,3]
plt.stackplot(x,area)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
area1=[6,2,4,5,3]
area2=[1,2,8,5,4]
area3=[4,6,1,2,3]
l=["area1","area2","area3"]
plt.stackplot(x,area1,area2,area3,labels=l,colors=["g","y","r"],)
#baseline="zero",sym,"wiggle
plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
# plt.grid()
plt.legend(loc=2)
plt.show()
```
