# Stem plot

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,4,2,3,1]
plt.stem(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,4,2,3,1]
plt.stem(x,y,linefmt=":",markerfmt="ro",bottom=0,basefmt="g",label="python",
        orientation="horizontal")
#use_line_collection=True
plt.legend()
plt.show()
```
