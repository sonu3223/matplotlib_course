# Box plot

```python
import matplotlib.pyplot as plt
x=[10,20,30,40,50,60,70]
plt.boxplot(x)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40,50,60,70,120]
plt.boxplot(x,labels=["python"],showmeans=True,sym="b*",boxprops=dict(color="r"),
            capprops=dict(color="b"),whiskerprops=dict(color="g"),))
#notch="True" vert=False  widths=.1  patch_artist=True whis=2 
#flierprops=dict(markeredgecolor="black"
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40,50,60,70]
x1=[10,80,40,90,50,70,60]
y=[x,x1]
plt.boxplot(y,labels=["python","c++"],showmeans=True,sym="b*",boxprops=dict(color="r"),capprops=dict(color="b"),whiskerprops=dict(color="g"))
plt.show()
```
