# Scatter plot

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[3,2,5,7,8,1]
plt.scatter(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[3,2,5,7,8,1]
colors=["r","g","b","y","black","b"]
sizes=[200,210,300,400,500,350]
plt.scatter(x,y,c=colors,s=sizes,marker="*",edgecolor="orange",linewidth=2) #,alpha=.3
plt.title("scatter",fontsize=15)
plt.xlabel("Days",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[3,2,5,7,8,1]
colors=[30,20,40,50,70,60]
sizes=[200,210,300,400,500,350]
plt.scatter(x,y,c=colors,s=sizes,cmap="viridis")
# plt.colorbar()
t=plt.colorbar()
t.set_label("colorbar",fontsize=15)
plt.title("scatter",fontsize=15)
plt.xlabel("Days",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[3,2,5,7,8,1]
y1=[4,7,6,9,2,3]
colors=[30,20,40,50,70,60]
sizes=[200,210,300,400,500,350]
plt.scatter(x,y,c=colors,s=sizes,cmap="viridis")
plt.scatter(x,y1,color="b",s=sizes)
t=plt.colorbar()
t.set_label("colorbar",fontsize=15)
plt.title("scatter",fontsize=15)
plt.xlabel("Days",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.show()
```
