```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,4,6,8,9]
plt.plot(x,y)
plt.xticks(x)
plt.yticks(y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[9,3,5,7]
plt.plot(x,y)
plt.xticks(x,labels=["python","c","c++","java"])
plt.yticks(y)
plt.show()
```
