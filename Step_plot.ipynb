# step plot

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[4,5,3,1,2]
plt.step(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[4,5,3,1,2]
plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()
plt.step(x,y,color="r",marker="o",ms=10,mfc="b",label="Stepplot")
plt.legend(loc=1)
plt.show()
```
